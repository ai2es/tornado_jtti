```python
preds_dir = "/ourdisk/hpc/ai2es/tornado/wofs-preds-2023/"
# dates = [i[-8:] for i in glob.glob(preds_dir+'2023****', recursive=True)]
dates = ["20230510", "20230511", "20230512", "20230523", "20230524", "20230526"]
dates
```

```python
init_dir = "/ourdisk/hpc/ai2es/tornado/wofs-preds-2023-init/2023/"
wofs_init_times = [i[-10:] for i in glob.glob(init_dir+dates[-1]+'/*', recursive=True)]
print(wofs_init_times)
```

```python
for date in dates[-1:]:

    init_times = sorted([i[-4:] for i in glob.glob(preds_dir+date+'/****', recursive=True)])
    wofs_init_times = sorted([i[-12:] for i in glob.glob(init_dir+date+'/*', recursive=True)])

    for init_time in init_times[7:9]:
        if init_time.startswith('0'):
            wofs_init_time = list(filter(lambda x: x.endswith(init_time), wofs_init_times))[0]
        else:
            wofs_init_time = date+init_time
        print(date, init_time, wofs_init_time)
        for ens in range(1, 19):
            filepath_wofs = init_dir + f"{date}/{wofs_init_time}/ENS_MEM_{ens}/wrfinput_d01_predictions.nc"
            wofs = xr.open_dataset(filepath_wofs, engine=engine, decode_times=False, decode_coords=True)
    
            files = sorted(glob.glob(preds_dir+f"{date}/{init_time}/ENS_MEM_{ens}/**.nc", recursive=True))
            for file in files[:2]:
                preds = xr.open_dataset(file, engine=engine, decode_times=False, decode_coords=True)
                print(np.isnan(preds['ML_PREDICTED_TOR'].values).sum())
                preds = preds.merge(wofs[["HGT", "P", "PB", "PH", "PHB"]], compat='override', join='override')
                # preds = preds.merge(wofs["HGT"], compat='override', join='override')
                # preds["HGT"] = (['Time', 'south_north', 'west_east'],  wofs["HGT"].data)
                # preds["P"] = (['Time', 'south_north', 'west_east'],  wofs["P"].data)
                # preds["PB"] = (['Time', 'south_north', 'west_east'],  wofs["PB"].data)
                # preds["PH"] = (['Time', 'south_north', 'west_east'],  wofs["PH"].data)
                # preds["PHB"] = (['Time', 'south_north', 'west_east'],  wofs["PHB"].data)
                print(np.isnan(preds['HGT'].values).sum())
    
                savepath = file.replace('wofs-preds-2023', 'wofs-preds-2023-hgt')
                os.makedirs(savepath.split('wrfwof')[0], exist_ok=True)
    
    encoding_vars = [k for k in preds.variables.keys() if k not in ["Times", "Time"]]
    encoding = {var: {"zlib":True, "complevel":4, "least_significant_digit":2} for var in encoding_vars}
    encoding["ML_PREDICTED_TOR"]['least_significant_digit'] = 3
    preds.to_netcdf(savepath, encoding=encoding)
```
